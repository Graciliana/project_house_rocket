## changes in the dataset

**Observo que não há NA ou NULL no data_set.**

- Modificar as colunas para Português.
- Modificar a coluna 2 (Data) para tipo data %Y-%M-%D - já tratada
- Modificar a coluna "waterfront" para string.
- Modificar as colunas "sqft" de int para float.
- Verificar se os ID duplicados fazem parte do fenômeno ou foi algum tipo de erro.
- Verificar a necessidade dessas duas colunas "sqft_living15" "sqft_lot15", caso desnecessáio para esta análise, descartar.
- Verificar a necessidade de mudança de pés quadrados "sqft" para m² "m2".
- Verificar o imóvel com 33 "bedrooms" se realmente há possibilidade de haver 33 quartos ou se pode ser um erro/outlier.
    - Comparar as casas com um ranged de 10% maior ou 10% menor de m², verificar quantos quartos tem essas casa.

```python
# remover colunas desnecessarias
df.drop(columns=['sqft_living15','sqft_lot15'], inplace=True)
```

```python
df.columns
```

```python
# Verificar quais linhas têm bedrooms igual a 33
df[df['bedrooms'] == 33]
```

```python
# substituir o valor bedrooms por 3 parece um erro de digitação
df.loc[df['bedrooms'] == 33, 'bedrooms'] = 3
```

```python
df.describe()
```

```python
# Modificar as colunas "sqft" de int para float.
# garantir que as colunas sqft_living, sqft_lot, sqft_above e sqft_basement são float
df['sqft_living'] = df['sqft_living'].astype(float)
df['sqft_lot'] = df['sqft_lot'].astype(float)
df['sqft_above'] = df['sqft_above'].astype(float)
df['sqft_basement'] = df['sqft_basement'].astype(float)
```

```python
# converter sqft para metros quadrados  pés quadrados para metros quadrados (1 ft² = 0.09290304 m²).
cols_to_convert = ['sqft_living', 'sqft_lot', 'sqft_basement', 'sqft_above']
for col in cols_to_convert:
    df[col] = df[col].apply(lambda x: round(x * 0.09290304, 2))
```

```python
# renomear as colunas 
df.rename(columns = {'id': 'id', 'date': 'data_venda', 'price':'preco', 'bedrooms':'quartos', 'bathrooms':'banheiros',
                          'sqft_living': 'm2_construido_total','sqft_lot':'m2_terreno_total', 'floors':'andares', 'waterfront':'beira-mar',
                          'view':'vista_geral', 'condition':'condicao', 'grade':'nivel_construcao','sqft_above':'m2_construidos_chao',
                          'sqft_basement':'m2_porao', 'yr_built':'ano_construído', 'yr_renovated':'ano_reformado', 'zipcode':'cep',
                          'lat':'latitude', 'long':'longitude'}, inplace=True)
```

```python
# verificar duplicidade de id e data_venda
df.loc[df['id'].duplicated(keep= False) == True].sort_values(['id','data_venda']).head(20)
```

```python
# verificar duplicadas
duplicatas = df[df.duplicated(subset=['id', 'data_venda'], keep=False)]
print(f"Número de duplicatas: {duplicatas.shape[0]}")
duplicatas
```
